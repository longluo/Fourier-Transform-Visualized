```python
from math import *
from scipy.integrate import quad
from scipy.signal import square

import numpy as np
import matplotlib.pyplot as plt
```

```python
wavelength = 200

x = np.arange(-500, 501, 1)
y = np.sin(2 * np.pi * x / wavelength)

plt.plot(x, y)
plt.show()
```

```python
SAMPLE_RATE = 44100  # Hertz
DURATION = 5  # Seconds

plt.style.use('seaborn-poster')

def generate_sine_wave(freq, sample_rate, duration):
    x = np.linspace(0, duration, sample_rate * duration, endpoint=False)
    frequencies = x * freq
    # 2pi because np.sin takes radians
    y = np.sin((2 * np.pi) * frequencies)
    return x, y

# Generate a 2 hertz sine wave that lasts for 5 seconds
x, y = generate_sine_wave(2, SAMPLE_RATE, DURATION)
plt.plot(x, y)
plt.xlabel('Time (x)')
plt.ylabel('Amplitude (y)')
plt.show()

x = np.linspace(0, 20, 201)
y = np.sin(x)

plt.figure(figsize=(8, 6))
plt.plot(x, y, 'b')
plt.xlabel('Location (x)')
plt.ylabel('Amplitude')

plt.show()

fig = plt.figure(figsize=(8, 8))

times = np.arange(5)

n = len(times)

for t in times:
    plt.subplot(n, 1, t + 1)
    y = np.sin(x + t)
    plt.plot(x, y, 'b')
    plt.plot(x[25], y[25], 'ro')
    plt.ylim(-1.1, 1.1)
    plt.ylabel('y')
    plt.title(f't = {t}')

plt.xlabel('location (x)')
plt.tight_layout()
plt.show()

# sampling rate
sr = 100.0
# sampling interval
ts = 1.0 / sr
t = np.arange(0, 1, ts)

# frequency of the signal
freq = 5
y = np.sin(2 * np.pi * freq * t)

plt.figure(figsize=(8, 8))
plt.subplot(211)
plt.plot(t, y, 'b')
plt.ylabel('Amplitude')

freq = 10
y = np.sin(2 * np.pi * freq * t)

plt.subplot(212)
plt.plot(t, y, 'b')
plt.ylabel('Amplitude')

plt.xlabel('Time (s)')
plt.show()

# frequency of the signal
freq = 5
y = 5 * np.sin(2 * np.pi * freq * t)

plt.figure(figsize=(8, 8))
plt.subplot(211)
plt.plot(t, y, 'b')
plt.ylabel('Amplitude')

y = 10 * np.sin(2 * np.pi * freq * t + 10)

plt.subplot(212)
plt.plot(t, y, 'b')
plt.ylabel('Amplitude')

plt.xlabel('Time (s)')
plt.show()
```

```python
# Square Wave

x = np.arange(-2 * np.pi, 2 * np.pi, 0.001)
# x axis has been chosen from –2π to +2π,
# value of 1 smallest square along x axis is 0.001

y = square(x)
# defining square wave function
# 𝑦 =−1, 𝑓𝑜𝑟 −𝜋 ≤ 𝑥 ≤ 0
# y =+1, 𝑓𝑜𝑟 0 ≤ 𝑥 ≤ 𝜋

plt.figure(figsize = (12, 6))

plt.subplot(211)
plt.plot(x, y)
plt.title("Square Wave")

# define fuction

fc = lambda x: square(x) * cos(i * x)  # i: dummy index
fs = lambda x: square(x) * sin(i * x)

n = 50  # max value of I, not taken infinity, better result with high value

An = []  # defining array
Bn = []

sum = 0

for i in range(n):
    an = quad(fc, -np.pi, np.pi)[0] * (1.0 / np.pi)
    An.append(an)

for i in range(n):
    bn = quad(fs, -np.pi, np.pi)[0] * (1.0 / np.pi)
    Bn.append(bn)  # putting value in array Bn

for i in range(n):
    if i == 0.0:
        sum = sum + An[i] / 2
    else:
        sum = sum + (An[i] * np.cos(i * x) + Bn[i] * np.sin(i * x))

plt.subplot(212)
plt.plot(x, sum, 'g')
plt.plot(x, y, 'r--')
plt.title("Fourier Series for Square Wave")

plt.show()

```
